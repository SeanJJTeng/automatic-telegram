```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.externals import joblib
from sklearn import tree

game_data = pd.read_csv('vgsales.csv')

game_data.head(18)
```

```python
#inspect data
game_data.shape
```

```python
game_data.describe()
#Problem with Year column
```

```python
game_data.isnull().sum()
```

```python
game_data = game_data.dropna(subset=['Year'], how = 'all')

game_data.shape
```

```python
X = game_data.drop(columns=['Name', 'Platform','Genre','Publisher','NA_Sales', 'EU_Sales', 'Other_Sales','JP_Sales', 'Global_Sales'])
y = game_data.drop(columns=['Rank','Name', 'Platform','Year','Genre','Publisher', 'EU_Sales', 'Other_Sales','JP_Sales', 'Global_Sales'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
#Inspect training sets
X_train
```

```python
y_train
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)


score = accuracy_score(y_test, predictions)
```

```python
score
```

```python
joblib.dump(model, 'NAsales_predictor.joblib')
```

```python
model = joblib.load('NAsales_predictor.joblib')
```
